```python
import pandas as pd
datos_ventas = pd.read_csv('/content/drive/MyDrive/Visualización/dtos_ejercicio_ventas.csv')
datos_ventas.head()
```

1. ¿Cómo ha evolucionado la cuota de mercado de cada
categoría a lo largo del tiempo?

El objetivo de esta gráfica es mostrar la evolución temporal de la cuota de mercado correspondiente a cada categoría.
Para su elaboración, se han utilizado los siguientes datos: una variable numérica de Año, una variable de Mes, una variable denominada ‘AMOUNT’ que representa la cuota de mercado, y una variable categórica ‘SUBBRAND’ que identifica cada categoría.
El gráfico está dirigido a mi profesor, como parte de la entrega del ejercicio.
La principal limitación a la que me enfrento es el tiempo disponible para completar y entregar el trabajo (1 semana). Además no cuento con limitantes físicos.
El gráfico que se elige es un diagrama de área porque es la mejor manera de representar la evolución de cada categoría comparando al igual cada una de ellas con el resto.

```python
import plotly.express as px

media_cuota_mensual_subbrand = datos_ventas.groupby(['YEAR', 'MONTH', 'SUBBRAND'])['AMOUNT'].mean().reset_index()
media_cuota_mensual_subbrand['TIME'] = pd.to_datetime(media_cuota_mensual_subbrand['YEAR'].astype(str) + '-' + media_cuota_mensual_subbrand['MONTH'].astype(str))

fig = px.area(media_cuota_mensual_subbrand,
              x='TIME',
              y='AMOUNT',
              color='SUBBRAND',
              title='Evolución Cuota según la Categoría')
fig.update_layout(xaxis_title='Tiempo (Año, Mes)',
                  yaxis_title='Cuota Media')
fig.show()
```

2. ¿Qué categoría tiene el comportamiento más estacional a lo
largo del año?

El objetivo de esta gráfica es mostrar la estacionalidad de la cuota de mercado correspondiente a cada categoría.
Para su elaboración, se han utilizado los siguientes datos: una variable numérica de Año, una variable de Mes, una variable denominada ‘AMOUNT’ que representa la cuota de mercado, y una variable categórica ‘SUBBRAND’ que identifica cada categoría.
El gráfico está dirigido a mi profesor, como parte de la entrega del ejercicio.
La principal limitación a la que me enfrento es el tiempo disponible para completar y entregar el trabajo.
El gráfico adecuado para este caso es un gráfico de líneas que compara en cada categoría la evolución mensual en cada año. Siendo así fácil detectar una estacionalidad de un año a otro.

```python
import plotly.express as px

media_mensual_categoria = datos_ventas.groupby(['YEAR', 'MONTH', 'SUBBRAND'])['AMOUNT'].mean().reset_index()

media_mensual_categoria['TIME'] = pd.to_datetime(media_mensual_categoria['YEAR'].astype(str) + '-' + media_mensual_categoria['MONTH'].astype(str))

fig = px.line(media_mensual_categoria,
              x='MONTH',
              y='AMOUNT',
              color='YEAR',
              facet_col='SUBBRAND',
              facet_col_wrap=3,
              title='Comparación de estacionalidad según la categoría)')

fig.update_layout(xaxis_title='Mes',
                  yaxis_title='Cuota media mensual',
                  height=800)
fig.update_yaxes(matches=None)
fig.show()
```
